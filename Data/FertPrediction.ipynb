```python
import numpy as np # linear algebra
import pandas as pd
```

```python
df = pd.read_csv("/content/FertilizerData.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
soil = pd.get_dummies(df["Soil Type"])
crop = pd.get_dummies(df["Crop Type"])
```

```python
df = pd.concat([df, soil, crop], axis=1)
df = df.drop(["Soil Type", "Crop Type"], axis=1)
df.head()
```

```python
X = df.drop("Fertilizer Name", axis=1)
y = df["Fertilizer Name"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
import pickle
pickle_out = open("Fertclassifier.pkl","wb")
pickle.dump(rf_model, pickle_out)
pickle_out.close()
```
